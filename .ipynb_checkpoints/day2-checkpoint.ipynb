# NLP phase 1

```python
text = "I am feeling very very happy today!!!....."
print("Orginal text :",text)
```

```python
#step 1 - convert text into lowercase 
#because happy != happy
#ML will consider time as differnce


text = text.lower()
print("Lowercase:",text)
```

```python
#email regets patterns (IMPORTANT)
#weak password regex(Important)
#step2 Remove punctuations
#we remove everything excpet letters and spaces
# using re - regular expression

import re 
text = re.sub(r'[^\w\s]','',text)
print("Without punctuation:",text)
#[^\w\s] - remove everything except letters and spaces
```

```python
#step - 3 tokenization (split sentence into words)
token = text.split()
print("Tokens :",token)
```

```python
#step-4 - remove stopwords - i,am,is,are,the,a ,an
stopwords = ["i","am","is","the","a","an","very"]

filtered_tokens = []
for word in token:
    if word not in stopwords:
        filtered_tokens.append(word)
        
print("after removing stopwords : ",filtered_tokens)
```

```python
def stem(word):
    if word.endswith("ing"):
        return word[:-3]
    return word
        
stemmed_words = []
for word in filtered_tokens:
    stemmed_words.append(stem(word))
print("after stemming: ",stemmed_words)
```

# NLP phase 2 words to vector

```python
#why vector ?
#because ML only understand numbers
#we must convert words into numeric form 

#bag of words(BOW)
```

```python
#text into the count
from sklearn.feature_extraction.text import CountVectorizer
```

```python
Sentences = {
    "i am good",
    "i am happy",
    "happy today",
    "i am sad"
}
```

```python
#vectorize 
vectorizer = CountVectorizer() # creating object
x = vectorizer.fit_transform(Sentences)
#learn vocabluary and convert sentences to numbers
```

```python
print("Vocab :",vectorizer.get_feature_names_out())
```

```python
print("Vectors : ")
print(x.toarray())
```

```python
#here:
#each row = sentence
#each column  = word
#                 am |good |happy | sad |today
#s1 : i am good  1  |  1  |  0   |  0  | 0
#s2 : i am sad   1  |  0  |  0   |  1  | 0

sentences = {
    "i am learning python and python is good",
    "python is a great programming to lean",
    "i also love machine leaning with python",
}
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(sentences)
print("vocab :",vectorizer.get_feature_names_out())
print("\nMatrix :")
print(X.toarray())
```

# TF - IDF - Ter, Frequency - Inverse Document 

```python
# bag of words treats all words equally 
#for example : I am leaning python, python and python 
#it will assign huge weight to python
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
sentences = {
    "i am learning python and python is good",
    "python is a great programming to lean",
    "i also love machine leaning with python",
}
```

```python
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(sentences)
#Internally it will:
#built Vocabulary 
#calculate tf
#calculate idf
#multiple tf x idf
#generate matrix
```

```python
print("vocab: ",vectorizer.get_feature_names_out())
```

```python
print("TF-IDF MAtrix")
print(X.toarray())
```
